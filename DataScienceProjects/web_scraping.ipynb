```python
#Class and tags
mbox=body.find_all(class_="mergedtoprow")
mbox
```

```python
ref=body.select(".references li span")
print(ref)
```

```python
for r in ref:
    print(r.get_text().strip())
```

```python
lang_link=body.find_all('a',class_="interlanguage-link-target")
print(lang_link)
```

```python
for item in lang_link:
    print(item.get_text().strip())
```

```python
for item in lang_link:
    lang=item.get_text().strip()
    link=str(item).split("href=\"")[1].split("\"")[0]
    print(lang_link.index(item)+1,lang,link)
```

```python
language=[]
links=[]
for jj in lang_link:
    language.append(jj.get_text().strip())
    linkmain=str(jj).split("href=\"")[1].split("\"")[0]
    links.append(linkmain)
```

```python
print(language)
```

```python
print(links)
```

```python
import pandas as pd
```

```python
lang_link_frame=pd.DataFrame(language,columns=["Languages"])
lang_link_frame['Links']=links
print(lang_link_frame)
```

```python
langlink3=pd.DataFrame(zip(language,links),columns=["Languages","Links"])
langlink3
```
